#### 1. Reproducing a plot
1. Use `numpy`'s `random` module to generate a random list of size 1000 of the following three categories: "CS", "Engineering", "Math" and "Physics"
  1.1 Use a seed of 42
2. Store the data in an array called my `students_affiliations`
3. use `matplotlib` to plot the data in `students_affiliations` to reproduce the following graph
 3.1 The plot uses "#FF5733" as color with an alpha (transparency) of 0.55.

* Your graph should look exactly like the one below

![](https://www.dropbox.com/s/ffmu62es6d7evh5/experience_fig.png?dl=1)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
students_affiliations = np.random.choice(["CS", "Engineering", "Math", "Physics"],1000)
students_data = pd.Series(students_affiliations)
cs_count = students_data.value_counts()["CS"]
eng_count = students_data.value_counts()["Engineering"]
math_count = students_data.value_counts()["Math"]
phys_count = students_data.value_counts()["Physics"]

df = pd.DataFrame({'Disciplines':["CS", "Engineering", "Math", "Physics"], 'Counts':[cs_count, eng_count, math_count, phys_count]})
sorted_df = df.sort_values(by='Counts', ascending = False)
plt.bar(sorted_df['Disciplines'], sorted_df['Counts'], color='#FF5733', alpha=0.55)
plt.title("Distribution of Students in the Program", color='green')
plt.xlabel("Disciplines")
plt.ylabel("Counts")
```

#### 2. Generating a scatter plot
* Generate random data to produce a scatter plot
  * You can use any function to generate the data
  * Instead of using circles as symbols, you should use squares.
    

```python
weight = np.random.randint(110, 200, 100)
height = np.random.randint(60, 74, 100)

plt.scatter(height, weight, marker='s')
plt.xlabel("Height")
plt.ylabel("Weight")
plt.title("Representation of participants Height and Weight")
```
